```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
path='selected_data_ISONE.csv'
df=pd.read_csv(path, parse_dates=True,  index_col='Date')
df.head(5)
```

```python
df.reset_index('Date',inplace=True)
df['temp_date']=df['Date']
```

```python
df.set_index('Date',inplace=True)
df.head()
```

```python
df['hour']=df['hour']-1
```

```python
df['Datetime']=df['temp_date'].astype(str)+' '+df['hour'].astype(str)+':00'
```

```python
df.head()
```

```python
df=df['2004-01-01':]
```

```python
df.head()
```

```python
column_names = ['Datetime','demand','temperature']
df = df.reindex(columns=column_names)
df
```

```python
df['Datetime']=pd.to_datetime(df["Datetime"])
df.head()
```

```python
print('Data Points= ',len(df))
```

```python
df=df.sort_index(ascending=True)
df.head(2)
```

```python
df.reset_index('Date',inplace=True)
del df['Date']
# df.set_index('Datetime',inplace=True)
df.head()
```

```python
df['Date'] = df['Datetime'].dt.normalize()
```

```python
df.head(2)
```

```python
holiday=pd.read_csv('US Holiday Dates (2004-2021).csv', parse_dates=['Date'],usecols=['Date','Holiday'])
holiday.head(1)
```

```python
duplicateRows =holiday[holiday.duplicated(['Date'])]
duplicateRows
```

```python
holiday.drop_duplicates(subset=['Date'], keep ='first',  inplace= True)
print('len = ',len(holiday))
```

```python
holiday.head(2)
```

```python
holiday['Date'] = holiday['Date'].dt.normalize()
```

```python
holiday.head(2)
```

```python
df.head(1)
```

```python
mergedf = pd.merge(df, holiday, on= 'Date', how="left")
mergedf['Holiday'] = mergedf['Holiday'].notnull().astype("int")
mergedf.head(1)
```

```python
#mergedf.drop(mergedf['year'], inplace=True)
column_names = ['Datetime','demand','temperature','Holiday']
mergedf = mergedf.reindex(columns=column_names)
mergedf.head(2)
```

```python
mergedf.to_csv('processed_ISONE_')
```
